### Distance between Clusters: Complete Linkage(farthest neighbor)
-  Maximum distance between the members of two clusters is called as **farthest neighbor**

- Caculate the distance between all the members and take the average of the members
-  Average distance between the all the  members of two clusters is called as **Average Linkage**
