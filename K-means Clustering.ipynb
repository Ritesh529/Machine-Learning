## Clusters:
- Predetermined number of non-overlapping clusters
-  Clusters are homogenous yet dissimilar to other clusters
- Need measure of within cluster similarity and between cluster similarity
- No Hierarchy.End product is final clustermemberships
- Useful for large datasets
- wcs should be less(distance between the members of the cluster should be less)
- bcs should be more(distance between the centers of the cluster should be more)


## How to choose the number of clusters
- Elbow method
